```python
#display notes details
#https://fr.trustpilot.com/review/cofidis.fr
# todo à revoir

tpPage = urlopen("https://fr.trustpilot.com/review/cofidis.fr")
soup = BeautifulSoup(tpPage, 'html.parser')

#tpNicknames = soup.findAll(name = 'h2', attrs = {'class': 'typography_appearance-default__AAY17'})
#tpComments = soup.findAll(name = 'p', attrs = {'class': 'typography_color-black__5LYEn'})
tpavis = soup.findAll(name = 'p', attrs = {'class': 'typography_body-l__KUYFJ typography_appearance-default__AAY17 typography_color-black__5LYEn'})
tptitlComments = soup.findAll(name = 'h2', attrs = {'class': 'typography_heading-s__f7029 typography_appearance-default__AAY17'})
tpdate =  soup.findAll(name = 'p', attrs = {'data-service-review-date-of-experience-typography' : 'true'})
#tptime1 = soup.findAll(name = 'time', attrs = {'data-service-review-date-time-ago' :"true"})
tptime = soup.findAll(name = 'time', attrs = {'class' :""})
tpuser = soup.findAll(name = 'span', attrs = {'data-consumer-name-typography' :"true"})
tprep = soup.findAll(name = 'p', attrs = {'data-service-review-business-reply-text-typography' :"true"})


titleComments = [] # 
for element in tptitlComments:
    titleComments.append(element.text.encode('ascii', errors='ignore').decode('utf-8'))

comment = [] # 
for element in tpavis:
    comment.append(element.text.encode('ascii', errors='ignore').decode('utf-8'))

tdate = [] # 
for element in tpdate:
    tdate.append(element.text.encode('ascii', errors='ignore').decode('utf-8'))
t_time = [] # 
for element in tptime:
    t_time.append(element.text.encode('ascii', errors='ignore').decode('utf-8'))



username = [] # 
for element in tpuser:
    username .append(element.text.strip("(text)"))

respcompany = [] # 
for element in tprep:
    respcompany .append(element.text.encode('ascii', errors='ignore').decode('utf-8'))

df_comm = pd.DataFrame(list(zip(comment, titleComments, tdate,t_time ,username,respcompany)), columns = ["Comments", "Titre de commande", "date", 'time', 'user','response of company'])

df_comm.head()
```

```python
import csv
a = df_comm.head(20)
a.to_csv(r'./avis1.csv', sep=';', encoding='utf-8', header='true')
```
